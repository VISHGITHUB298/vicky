## Use pivot table method to identify dates on which maximum sale occurred?

```python
import numpy as np
df6 = pd.read_csv('sales (1).csv')
print('Original DataFrame :\n')
print(df6)
table = pd.pivot_table(data=df6, index=['ORDERDATE'] , values='AMOUNTINFO', aggfunc=np.max)
print('\n\nPivot table with dates on which maximum sales occured:\n')
table
```

```python
import pandas as pd 
import numpy as np 

from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],
             columns = np.append(cancer_dataset['feature_names'], ['target']))
```

```python
cancer_df.info()
```

```python
cancer_df.describe()
```

```python
X = cancer_df.drop(['target'], axis = 1)
y = cancer_df['target']
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= 5)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
from sklearn.svm import SVC
svc_classifier2 = SVC()
svc_classifier2.fit(X_train_sc, y_train)
y_pred_svc_sc = svc_classifier2.predict(X_test_sc)
accuracy_score(y_test, y_pred_svc_sc)
     
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators = 20, criterion = 'entropy', random_state = 51)
rf_classifier.fit(X_train, y_train)
y_pred_rf = rf_classifier.predict(X_test)
accuracy_score(y_test, y_pred_rf)
```
