```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales_data = pd.read_csv("Sales.csv")
sales_data
```

```python
month = sales_data['month_name'].tolist()
print(month)

profit = sales_data['total_profit'].tolist()
print(profit)

plt.plot(month, profit, color='b', ls=':', marker='o', markerfacecolor='y', linewidth='5')

plt.xlabel('Month Number')
plt.ylabel('Item Sold units number')

plt.legend(['Profit'], loc ="upper right")
plt.show()
```

```python
plt.plot(month, profit, linestyle= '--' , color='green', linewidth = 3 , marker = 'o', mfc='k' ,label ="Profit data of last year",)
plt.xlabel("Month Number")
plt.ylabel("Sold units nuber")
plt.title("Company Sales Data of Last Year")
plt.legend(loc ="upper right")
plt.ylim(100000)
plt.xlim()
plt.show()
```
