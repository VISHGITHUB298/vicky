```python
#Q3). Create a JSON file for below object and Create a Pandas DataFrame using the created JSON file?
# {"Courses":{"r1":"Spark"},"Fee":{"r1":"25000"},"Duration":{"r1":"50 Days"}} 
```

```python
import json

data = {"Courses": {"r1": "Spark"}, "Fee": {"r1": "25000"}, "Duration": {"r1": "50 Days"}}

with open('data.json', 'w') as outfile:
    json.dump(data, outfile)
```

```python
import pandas as pd

with open('data.json') as json_file:
    data = json.load(json_file)

df = pd.json_normalize(data)

print(df)
```

```python
#Q7)Create a DataFrame using below link:
#https://raw.githubusercontent.com/ffzs/dataset/master/sales.csv
#Use pivot table method to identify dates on which maximum sale occurred
```

```python
import pandas as pd
df=pd.read_csv('salessorder.csv')
df
```

```python
import pandas as pd

df = pd.read_csv('salessorder.csv')

df['date'] = pd.to_datetime(df['ORDERDATE'])

max_sales = pd.pivot_table(df, values='AMOUNTINFO', index='ORDERDATE', aggfunc='max')

max_sales_dates = max_sales[max_sales['AMOUNTINFO'] == max_sales['AMOUNTINFO'].max()].index

print(max_sales_dates)
```
