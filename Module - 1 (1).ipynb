```python
import matplotlib.pyplot as plt

y1 = [14, 67, 23, 89]

y2 = [23, 78, 12, 34]

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].plot(y1)
axs[0].set_title('Plot 1')

axs[1].plot(y2)
axs[1].set_title('Plot 2')

for ax in axs.flat:
    ax.set(xlabel='X-axis', ylabel='Y-axis')

plt.show()
```

# Q-21

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
sales_data = pd.read_csv('sales.csv')
```

```python
plt.plot(sales_data['Month'], sales_data['Total Profit'])
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit of All Months')
plt.show()

# Total profit of all months with style properties
plt.plot(sales_data['Month'], sales_data['Total Profit'], linestyle='dotted', color='blue', linewidth=5, marker='o', markerfacecolor='yellow')
plt.xlabel('Month Number')
plt.ylabel('Item Sold Units Number')
plt.title('Total Profit of All Months with Style Properties')
plt.legend(loc='upper right')
plt.show()

# Total units sale data for last year for each product using pie chart
labels = sales_data['Product']
sizes = sales_data['Units Sold']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Total Units Sold for Each Product')
plt.show()
```

# Q-20

```python
import numpy as np
import matplotlib.pyplot as plt

amplitude = 4
frequency = 50

x = np.linspace(0, 2*np.pi, 1000)

y = amplitude * np.cos(frequency * x)

plt.plot(x, y)

plt.title("Cosine Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```
