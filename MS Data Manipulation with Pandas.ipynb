```python
#Q.3. Create a JSON file for below object and Create a Pandas DataFrame using the created JSON file?
 #         {"Courses":{"r1":"Spark"},"Fee":{"r1":"25000"},"Duration":{"r1":"50 Days"}} 
```

```python
import json

data = {"Courses": {"r1": "Spark"}, "Fee": {"r1": "25000"}, "Duration": {"r1": "50 Days"}}

with open('data.json', 'w') as outfile:
    json.dump(data, outfile)
```

```python
import pandas as pd

with open('data.json') as json_file:
    data = json.load(json_file)

df = pd.json_normalize(data)

print(df)
```

```python
#Q.5.For the below dictionary, create a Pandas DataFrame and select all rows with Percentage >=80.
#record = { 
# 'Name': ['Ankit', 'Amit', 'Aishwarya', 'Priyanka', 'Priya', 'Shaurya' ],
# 'Age': [21, 19, 20, 18, 17, 21],
# 'Stream': ['Math', 'Commerce', 'Science', 'Math', 'Math', 'Science'],
# 'Percentage': [88, 92, 95, 70, 65, 78] } 
```

```python
import pandas as pd

record = {
    'Name': ['Ankit', 'Amit', 'Aishwarya', 'Priyanka', 'Priya', 'Shaurya'],
    'Age': [21, 19, 20, 18, 17, 21],
    'Stream': ['Math', 'Commerce', 'Science', 'Math', 'Math', 'Science'],
    'Percentage': [88, 92, 95, 70, 65, 78]
}

df = pd.DataFrame(record)
df_filtered = df[df['Percentage'] >= 80]

print(df_filtered)
```

```python
import pandas as pd

df = pd.read_csv('salessorder.csv')

df['date'] = pd.to_datetime(df['ORDERDATE'])

max_sales = pd.pivot_table(df, values='AMOUNTINFO', index='ORDERDATE', aggfunc='max')

max_sales_dates = max_sales[max_sales['AMOUNTINFO'] == max_sales['AMOUNTINFO'].max()].index

print(max_sales_dates)
```
