# 1. Write a Python program to swap first and last element of the list.

```python
#swap first and last element in list

# Swap function
def swapList(sl):
    n = len(sl)
      
    # Swapping 
    temp = sl[0]
    sl[0] = sl[n - 1]
    sl[n - 1] = temp
      
    return sl
      
l = [13, 14, 5, 9, 56, 33]

print(l)
print("Swapped list: ",swapList(l))
```

# 6. Write a python program to find whether given string is palindrome.
## eg: amaama

```python
str = "amaama"
l = len(str)
j = (int)(len(str)/2)
fl = True
for i in range(j):
    if(str[i] != str[l-i-1]):
        fl = False
        break
    
if(fl == True): 
    print("Given string is palindrome.")
else:
    print("Given string is not palindrome.")
```

# 14. Write a python program to add matrices of dimension 3-by-3 with given matrix
a=[[ 9 10 11]
[12 13 14]
[15 16 17]]
b=[[ 19 30 11]
[12 13 44]
[15 56 17]]

```python
X = [ [ 9, 10, 11],
    [12 ,13, 14],
   [15, 16 ,17]]
 
Y = [[ 19, 30, 11],
     [12, 13, 44],
     [15, 56 ,17]]
 
 
result = [[0,0,0],
        [0,0,0],
        [0,0,0]]
 
# iterate through rows
for i in range(len(X)):  
# iterate through columns
    for j in range(len(X[0])):
        result[i][j] = X[i][j] + Y[i][j]
 
for r in result:
    print(r)
```

# 20. Write a python Matplotlib program to plot cosine graph with below parametric values:
Amplitude: 4
frequency = 50

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the amplitude and frequency
amplitude = 4
frequency = 50

# Define the x-axis values
x = np.linspace(0, 2*np.pi, 1000)

# Calculate the y-axis values
y = amplitude * np.cos(frequency * x)

# Plot the graph
plt.plot(x, y)

# Add a title and axis labels
plt.title("Cosine Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Show the graph
plt.show()
```

# 24. Write a python Matplotlib program using sub plotting to generate line plots for values:
For plot1: y1=[14,67,23,89]
For plot2: y2=[23,78,12,34]

```python
import matplotlib.pyplot as plt

y1 = [14, 67, 23, 89]

y2 = [23, 78, 12, 34]

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].plot(y1)
axs[0].set_title('Plot 1')

axs[1].plot(y2)
axs[1].set_title('Plot 2')

for ax in axs.flat:
    ax.set(xlabel='X-axis', ylabel='Y-axis')

plt.show()
```
